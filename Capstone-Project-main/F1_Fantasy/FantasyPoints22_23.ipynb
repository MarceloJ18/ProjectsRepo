```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver import Firefox
import re
import json
import random

```

```python
# Make HTML header for browser
# User agent for Firefox with Geko driver

user_agent_browser = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:121.0) Gecko/20100101 Firefox/121.0'
# All the following options are not necessary, but they are good practice
# They can be found in the Network tab of the Developer Tools, in the Headers tab
accept_decoding = "gzip, deflate, br"
accept_language = "pt-PT,pt;q=0.8,en;q=0.5,en-US;q=0.3"
accept = 'application/json, text/javascript, */*; q=0.01'
accept_textHTML_only = 'text/html;q=0.9' # specific file wanted


options = webdriver.FirefoxOptions()
options.add_argument("--headless") # Set the Firefox webdriver to run in headless mode for scalability
options.add_argument(f'user-agent={user_agent_browser}')
options.add_argument(f'accept-encoding={accept_decoding}')
options.add_argument(f'accept-language={accept_language}')
options.add_argument(f'accept={accept}')
```

```python

# By default, Selenium waits for all resources to download before taking actions.
# However, we don't need it as the page is populated with dynamically generated JavaScript code.
options.page_load_strategy = "none"
# Get the path of the geckodriver
geckodriver_path = "C:\\Users\\marce\\Downloads\\geckodriver-v0.33.0-win64"
geckodriver_path_full = "C:\\Users\\marce\\Downloads\\geckodriver-v0.33.0-win64\\geckodriver"

# Pass the defined options objects to initialise the web driver
driver = Firefox(options=options, executable_path=geckodriver_path_full)

# Set an implicit wait of 5 seconds to allow time for elements to appear before throwing an exception
driver.implicitly_wait(5)
```

```python
url = "https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/driver?season=2023"

driver.get(url)
time.sleep(15)
```

```python
# Get the page source and close the browser
page_source = driver.page_source
driver.close()
```

```python
# Create function to simplify all that was done
def retrieve_from_html(page_source, file_name):
    # Remove html contents such as <html><head></head><body>
    page_source_clean = re.sub('<[^<]+?>', '', page_source)
    # Convert the string to a list of dictionaries
    data_list = json.loads(page_source_clean)
    # Export data_list json to a file
    with open(f"C:\\Users\\marce\\Downloads\\F1_Data\\F1_{file_name}_list.json", "w") as f:
        json.dump(data_list, f)
    # Convert the list of dictionaries to a dataframe
    df = pd.json_normalize(data_list)
    return df
```

# Remaining Data


```python
page_url_drivers23 = 'https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/driver?season=2023'
page_url_constructors23 = "https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/constructor?season=2023"
page_url_races23 = 'https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/races?season=2023'

# 2022 data
page_url_drivers22 = 'https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/driver?season=2022'
page_url_constructors22 = "https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/constructor?season=2022"
page_url_races22 = 'https://f1fantasytoolsapi-szumjzgxfa-ew.a.run.app/race-results/races?season=2022'

pages_url = [page_url_drivers23, page_url_constructors23, page_url_races23, page_url_drivers22, page_url_constructors22, page_url_races22]
data_names = ['drivers23', 'constructors23', 'races23', 'drivers22', 'constructors22', 'races22']
```

```python

for i in range(len(pages_url)):
    driver = Firefox(options=options, executable_path=geckodriver_path_full)
    driver.get(pages_url[i])
    time.sleep(random.randint(15, 35))
    page_source = driver.page_source
    driver.close()
    df = retrieve_from_html(page_source, data_names[i])
    globals()[f'{data_names[i]}'] = df
    print(f'{data_names[i]} created! Pages left: {len(pages_url)-i-1}.\n')

```

```python
# Export dataframes to csv files
constructors23.to_csv("C:\\Users\\marce\\Downloads\\F1_23_constructors_basic.csv", index=False)
races23.to_csv("C:\\Users\\marce\\Downloads\\F1_23_races_basic.csv", index=False)
drivers22.to_csv("C:\\Users\\marce\\Downloads\\F1_22_drivers_basic.csv", index=False)
constructors22.to_csv("C:\\Users\\marce\\Downloads\\F1_22_constructors_basic.csv", index=False)
races22.to_csv("C:\\Users\\marce\\Downloads\\F1_22_races_basic.csv", index=False)
```
